```python
#读取数据
path=r'D:\课件\大三暑期\投资学\数据.csv'
df=pd.read_csv(path,header=0)
print(df)
```

```python
#按证券代码分组
for name,group in df.groupby("code"):
    print(name)
    print(group)
```

```python
#按组筛选数据
def df_code(i):
    code_i=df[df['code']==i]
    return code_i

df_code(145)
```

```python
#输出某组某项数据
def df_select(i,p):
    code_i=df[df['code']==i]
    select_i=code_i[p]
    return select_i

df_select(44,['price','Date'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request as urllib2
import json

price = []
for c in code:
    df_p=df_select(c,'price')
    price.append(np.array(df_p))
price = np.array(price)
price = price.T
df = pd.DataFrame(data=price, columns=code)
df.plot()
plt.show()
```

```python
import seaborn as sns
sns.heatmap(df.corr(), annot=True, square=True)
plt.show()
```

```python
rt = []
for c in code:
    df_r=df_select(c,'return')
    rt.append(np.array(df_r))
rt = np.array(rt)
rt = rt.T
df = pd.DataFrame(data=rt, columns=code)
df.plot()
plt.show()
```

```python
import seaborn as sns
sns.heatmap(df.corr(), annot=True, square=True)
plt.show()
```
